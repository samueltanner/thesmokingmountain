```python
def download_images():
    output_dir = "/Users/samtanner/Documents/Code/smoking_mtn/public/images/slide_thumbnails"
    os.makedirs(output_dir, exist_ok=True)

    for image_url in presentation_images:
        # Extract file ID from the Google Drive URL
        file_id = image_url.split('https://drive.google.com/u/0/drive-usercontent/')[1].split('=w200-h190')[0]
        
        # Construct direct download URL
        download_url = image_url
        
        # Create filename using file ID
        image_path = os.path.join(output_dir, f"{file_id}.jpg")
        
        try:
            response = requests.get(download_url)
            response.raise_for_status()
            
            with open(image_path, 'wb') as f:
                f.write(response.content)
            print(f"Downloaded: {image_path}")
            
        except Exception as e:
            print(f"Error downloading {file_id}: {str(e)}")
```
